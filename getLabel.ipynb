```python
def listdir(path, list_name):  
    for file in os.listdir(path):
        file_path = os.path.join(path, file)
        if os.path.isdir(file_path):
            listdir(file_path, list_name)
        else:
            list_name.append(file_path)
 


def get_path(labels,path,te,op):

#     path is the dirc path (eg. path='E:\dataset\DAiSEE\DAiSEE\csv_file\\test')
#     test is for test or train or validation, is the last characteristic of path
#     openFace/openPose/avi is the column that wait to be changed
    print(te,op)
    # test = str(test)
    # openFace = str(openFace)
    idx = labels['ClipID']
    list_name = []
    listdir(path,list_name) # get the abs file path in path
    for n, i in enumerate(list_name):
      p = list_name[n].index(te)
      t = list_name[n][p + 1+len(te):-4] + str('.avi')
      tm = list_name[n][p + 1+len(te):-4] + str('.mp4')

      if t in list(idx):
        colum = int(list(idx).index(t))
        labels.loc[colum,op] = list_name[n]

      if tm in list(idx):
        colum = int(list(idx).index(tm))
        if colum == 3124:
          print(tm)
        labels.loc[colum,op] = list_name[n]



def get_path_avi(labels,path,te,op):

#     path is the dirc path (eg. path='E:\dataset\DAiSEE\DAiSEE\csv_file\\test')
#     test is for test or train or validation, is the last characteristic of path
#     openFace/openPose/avi is the column that wait to be changed
    print(te,op)
    # test = str(test)
    # openFace = str(openFace)
    idx = labels['ClipID']
    list_name = []
    listdir(path,list_name) # get the abs file path in path
    for n, i in enumerate(list_name):
      p = list_name[n].index(te)
      t1 = list_name[n][p + 1+len(te):]
      p1 = t1.index('/')
      t2 = t1[p1+1:]
      p2 = t2.index('/')
      t = t2[p2 + 1:-4] + str('.avi')
      tm = t2[p2 + 1:-4] + str('.mp4')
        
      if n == 2:
        print(t)
        print(tm)

      if t in list(idx):
        colum = int(list(idx).index(t))
        labels.loc[colum,op] = list_name[n]

      if tm in list(idx):
        colum = int(list(idx).index(tm))
        labels.loc[colum,op] = list_name[n]
```

```python
# avi + test
path='/content/data/avi/Test'   
te = 'Test'
op = 'avi'
get_path_avi(label,path,te,op)

# avi + train
path='/content/data/avi/Train'   
te = 'Train'
op = 'avi'
get_path_avi(label,path,te,op)

# avi + val
path='/content/data/avi/Validation'   
te = 'Validation'
op = 'avi'
get_path_avi(label,path,te,op)
```
